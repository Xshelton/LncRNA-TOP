## generate the features from sequence csv files
### the csv file should be like this:


| label    | sequence |
| ----------- | ----------- |
| ACE2      | TCGA...       |
| AlBG   | AGCGC...        |

can be applied to generate either gene or lncRNA sequence's features .

```python

#Author: weidunxie2-c@cityu.edu.hk

import pandas as pd
df=pd.read_csv('gene_max_sequence16127.csv')
try:
 mirna=df['label']
 mirna_seq=df['sequence']
except:
 mirna=df['reference']
 mirna_seq=df['seq']   
    
def covert(seq):
    key=''
    rseq=seq[::-1]
    #print(rseq)
    for i,ap in enumerate(rseq):
        if ap=='A':
            key=key+'T'
        if ap=='T':
            key=key+'A'
        if ap=='C':
            key=key+'G'
        if ap=='G':
            key=key+'C'
    #print('rsq',rseq,'key',key)
    return key
    #for i in range(0,seq):
    
def count_plalindrome(seq,n):#give a sequence, and a number, calculate all the plalindrome it has
    list_pla=[]
    list_loc=[]#record_location
    list_order=[]
    list_count=[]
    list_key=[]
    list_win=[]
    for i in range(n,len(seq)+1-n):
        win=seq[i]
        #print('cursor',win)
        for j in range(1,n):
            win=win+seq[i+j]
        #print('real windows',win)
        #key=covert(win)
        if len(win)==n and len(win)+n<=len(seq):#this means generate all of the windows
            list_win.append(win)
    #print(list_win)
    for i in range(0,len(seq)+1-n):#GO through all of the windows
        win=seq[i:i+n]
        #print('find win',win)
        key=covert(win)# find out the plalindorm seq,
        if key in list_win:#if the key in all of the windows ,means that plalindorm exist
           if i<=list_win.index(key)+n:
              pla_group='{}-{}'.format(win,key)
              list_pla.append(pla_group)
              list_loc.append('{}-{}'.format(i,list_win.index(key)+n))
              if pla_group not in list_order:
                 list_order.append(pla_group)
                 list_count.append({'{}_count'.format(pla_group):1})
              else:
                 order_index=list_order.index(pla_group)
                 list_count[order_index]['{}_count'.format(pla_group)]=list_count[order_index]['{}_count'.format(pla_group)]+1
    if len(list_loc) !=0:
       for kk,key in enumerate(list_pla):
          key2={'pla':key,'location':list_loc[kk]}
          list_key.append(key2)
    print(list_count)
    #print(list_key)
    return list_key
ls=count_plalindrome(mirna_seq[2],2)
print(ls)
#for kk in range(1,2):
# ls_all=[]
# for i in range(0,15):
#  ls=count_plalindrome(mirna_seq[2],20-i)
#  if len(ls)!=0:
#    ls_all=ls_all+ls
#  print('for this length',20-i,'we have in total',len(ls))
#  print('ls_all',len(ls_all))
#print(ls_all)
```

```python
key_set={}
upper=len(df)#generally1 len(df)
number=7
for pla_num in range(number,number+1):
    #pla_num=5
    import itertools
    for i in itertools.product('TCGA', repeat = pla_num):
        #print(i)
        obj=''.join(i)
        # print(obj)
        ky={'{}'.format(obj):0}
        key_set.update(ky)
    print('finishing generating key_set')

    for i in range(0,upper):
      if i%1000==0 and i!=0:
        print('pla_num',pla_num,'schedule',i)
      key_set=clean_key_set(key_set)
      ls,lembed=count_plalindrome_pg(mirna_seq[i],pla_num,key_set)
      if i==0:
        lkset=pd.DataFrame(lembed)
      else:
        temp=pd.DataFrame(lembed)
        lkset=pd.concat([lkset,temp],axis=0)
      #print(lkset.shape)  
    lkset['label']=mirna[0:upper].tolist()
    lkset=lkset.loc[:, (lkset != 0).any(axis=0)]
    lkset.to_csv('new_palindromic_lnc_12417_count_pla{}.csv'.format(pla_num),index=None,)
    print('saving the plaindrom gene of ',pla_num)
    #print(ls)
```
